```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Cargar el conjunto de datos Digits
digits = load_digits()

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=42)

# Crear los modelos de clasificación
knn = KNeighborsClassifier(n_neighbors=5)
svm = SVC(kernel='linear')
tree = DecisionTreeClassifier()
forest = RandomForestClassifier(n_estimators=100)
logistic = LogisticRegression(max_iter=1000)

# Entrenar los modelos
knn.fit(X_train, y_train)
svm.fit(X_train, y_train)
tree.fit(X_train, y_train)
forest.fit(X_train, y_train)
logistic.fit(X_train, y_train)

# Hacer predicciones en el conjunto de prueba
y_pred_knn = knn.predict(X_test)
y_pred_svm = svm.predict(X_test)
y_pred_tree = tree.predict(X_test)
y_pred_forest = forest.predict(X_test)
y_pred_logistic = logistic.predict(X_test)

# Mostrar el reporte de clasificación y la matriz de confusión para cada modelo
print("KNN:")
print(classification_report(y_test, y_pred_knn))
print(confusion_matrix(y_test, y_pred_knn))

print("\nMáquina de Soporte Vectorial:")
print(classification_report(y_test, y_pred_svm))
print(confusion_matrix(y_test, y_pred_svm))

print("\nÁrbol de Decisión:")
print(classification_report(y_test, y_pred_tree))
print(confusion_matrix(y_test, y_pred_tree))

print("\nBosque Aleatorio:")
print(classification_report(y_test, y_pred_forest))
print(confusion_matrix(y_test, y_pred_forest))

print("\nRegresión Logística:")
print(classification_report(y_test, y_pred_logistic))
print(confusion_matrix(y_test, y_pred_logistic))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Cargar el conjunto de datos Digits
digits = load_digits()

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=42)

# Crear los modelos de clasificación
knn = KNeighborsClassifier(n_neighbors=5)
svm = SVC(kernel='linear')
tree = DecisionTreeClassifier()
forest = RandomForestClassifier(n_estimators=100)
logistic = LogisticRegression(max_iter=1000)

# Entrenar los modelos
knn.fit(X_train, y_train)
svm.fit(X_train, y_train)
tree.fit(X_train, y_train)
forest.fit(X_train, y_train)
logistic.fit(X_train, y_train)

# Hacer predicciones en el conjunto de prueba
y_pred = logistic.predict(X_test)  # Elegir el modelo deseado (ej. logistic)

# Calcular la matriz de confusión
cm = confusion_matrix(y_test, y_pred)

# Crear una figura y un eje
fig, ax = plt.subplots(figsize=(8, 8))

# Mostrar la matriz de confusión con colores
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax)

# Configurar las etiquetas de los ejes y el título
ax.set_xlabel('Clase Predicha')
ax.set_ylabel('Clase Real')
ax.set_title('Matriz de Confusión')

# Mostrar la gráfica
plt.show()
```
