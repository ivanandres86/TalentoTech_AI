```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# List of names
value_name_list = ["Byron", "Joel", "Diego", "Rocio", "Nykolas", "Julio", "Marcus", "Milo", "Rodolfo", "Mauricio"]

# List of gender
value_gender_list = ["M", "M", "M", "F", "M", "M", "M", "M", "M", "M"]

# List of ages
value_age_list = [25, 30, 35, 40, 28, 32, 29, 27, 31, 26]

# List of weights (in kg)
value_weight_list = [60.5, 75.0, 80.0, 90.5, 55.0, 70.0, 68.0, 72.5, 85.0, 65.0]

# List of heights (in cm)
value_height_list = [165, 180, 175, 190, 160, 172, 158, 185, 177, 162]

# List of years of experience
value_experience_list = [2, 5, 10, 15, 3, 6, 4, 1, 8, 9]

# List of salaries (in COP)
value_salary_list = [3000, 4500, 5000, 6000, 2800, 3200, 3800, 4000, 4200, 3500]

# List of number of children
value_children_list = [1, 0, 2, 3, 1, 2, 1, 0, 4, 2]

# Create a dictionary to represent all the information
data_dict = {
    "Name": value_name_list,
    "Gender" : value_gender_list,
    "Age": value_age_list,
    "Weight": value_weight_list,
    "Height": value_height_list,
    "Years_of_Experience": value_experience_list,
    "Salarie": value_salary_list,
    "Number_of_Children": value_children_list
}
```

```python
# Turn a dictionary into a dataframe
df = pd.DataFrame(data_dict)

df
```

```python
# Generate 100 additional data
num_additional_records = 100

# Create data random
additional_names = [f"Name_{i}" for i in range(len(value_name_list), len(value_name_list) + num_additional_records)]
additional_genders = np.random.choice(['M', 'F'], size=num_additional_records)
additional_ages = np.random.randint(20, 60, size=num_additional_records)
additional_weights = np.round(np.random.uniform(50, 100, size=num_additional_records), 1)
additional_heights = np.random.randint(150, 200, size=num_additional_records)
additional_experience = np.random.randint(0, 20, size=num_additional_records)
additional_salaries = np.random.randint(2500, 7000, size=num_additional_records)
additional_children = np.random.randint(0, 5, size=num_additional_records)

# add the new records directly to the existing dictionary
data_dict['Name'] += additional_names
data_dict['Gender'] += additional_genders.tolist()
data_dict['Age'] += additional_ages.tolist()
data_dict['Weight'] += additional_weights.tolist()
data_dict['Height'] += additional_heights.tolist()
data_dict['Years_of_Experience'] += additional_experience.tolist()
data_dict['Salarie'] += additional_salaries.tolist()
data_dict['Number_of_Children'] += additional_children.tolist()

df = pd.DataFrame(data_dict)
```

```python
df = df.loc[:, ~df.T.duplicated()]
df
```

```python
# Definir la función que acepta los parámetros 'Salarie' y 'Number_of_Children'
def request_salary_increase(row):
    if row['Number_of_Children'] < 2 and row['Salarie'] < 3000:
        return 'Solicitar aumento'
    else:
        return 'Estable'

# Usar .loc[] para aplicar la función y evitar la advertencia
df.loc[:, 'Request_increase'] = df.apply(request_salary_increase, axis=1)

df
```

```python
# Filtrar solo las columnas numéricas para la correlación
numerical_columns = df.select_dtypes(include=[np.number])

# Calcular la matriz de correlación solo para las columnas numéricas
correlation_matrix = numerical_columns.corr()

# Mostrar la matriz de correlación
print(correlation_matrix)
```
