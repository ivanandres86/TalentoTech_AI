```python
import random

# Define las entradas del perceptrón.
entradas = [random.randint(0, 1) for _ in range(5)]

# Define los pesos del perceptrón.
pesos = [random.uniform(-1, 1) for _ in range(5)]

# Define el umbral de activación.
umbral = 0.5

# Define la función de activación.
def funcion_activacion(suma_ponderada):
    """Devuelve 1 si la suma ponderada es mayor o igual al umbral, de lo contrario devuelve 0."""
    if suma_ponderada >= umbral:
        return 1
    else:
        return 0

# Calcula la suma ponderada de las entradas y los pesos.
suma_ponderada = sum([entradas[i] * pesos[i] for i in range(len(entradas))])

# Aplica la función de activación a la suma ponderada para obtener la salida.
salida = funcion_activacion(suma_ponderada)

# Imprime los datos.
print("Entradas:", entradas)
print("Pesos:", pesos)
print("Umbral:", umbral)
print("Suma ponderada:", suma_ponderada)
print("Salida:", salida)
```

```python
import random

# Define las entradas del perceptrón.
entradas = [[random.randint(0, 1) for _ in range(5)] for _ in range(10)]  # 10 sets of inputs

# Define los pesos del perceptrón.
pesos = [random.uniform(-1, 1) for _ in range(5)]

# Define el umbral de activación.
umbral = 0.5

# Define la función de activación.
def funcion_activacion(suma_ponderada):
    """Devuelve 1 si la suma ponderada es mayor o igual al umbral, de lo contrario devuelve 0."""
    if suma_ponderada >= umbral:
        return 1
    else:
        return 0

# Procesa cada entrada de datos
for entrada in entradas:
    # Calcula la suma ponderada de las entradas y los pesos.
    suma_ponderada = sum([entrada[i] * pesos[i] for i in range(len(entrada))])

    # Aplica la función de activación a la suma ponderada para obtener la salida.
    salida = funcion_activacion(suma_ponderada)

    # Imprime los datos.
    print("Entradas:", entrada)
    print("Pesos:", pesos)
    print("Umbral:", umbral)
    print("Suma ponderada:", suma_ponderada)
    print("Salida:", salida)
    print("-" * 20)  # Separador entre cada conjunto de datos
```
